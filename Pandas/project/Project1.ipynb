# Feature Extraction 


```python

import numpy as np
import pandas as pd
df=pd.read_csv('anime.csv')
```

```python
df.head()
```

```python
# name a new column on episode count
df.loc[0]['Title']
```

```python
def extract_ep(title):
    check = False
    data = ""

    for i in title:
        if i == ')':
            return data
        if i == '(':
            check = True
            continue   
        if check:
            data += i
    return None  
```

```python
df['Episode'] = df['Title'].apply(extract_ep)
```

```python
df['Episode']=df['Episode'].str.replace(" eps","")
```

```python
df['Episode']=df['Episode'].astype(int)
```

```python
#make a new col for timeStamp
def extract_time(title):
    check = False
    data = ""

    for i in range(len(title)):
        if title[i] == ')':
            for j in range(i+1,i+20):
                data+=title[j]
            return data
```

```python
df['Total Time'] = df['Title'].apply(extract_time)
```

```python
df['Total Time']
```

```python
def calc_months(x):
    start, end = x.split(' - ')
    
    start_date = pd.to_datetime(start)
    end_date = pd.to_datetime(end)
    
    return (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1
df['Total Months'] = df['Total Time'].apply(calc_months)
```
