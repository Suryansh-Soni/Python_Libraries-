## Group BY


```python
data = {
    "category": ["Electronics", "Clothing", "Groceries", "Electronics", "Electronics"],
    "store": ["Store A", "Store B", "Store C", "Store A", "Store B"],
    "sales": [15000, 8000, 3000, 22000, 12000],
    "quantity": [5, 20, 50, 8, 25],
    "date": [
        "2026-03-01",
        "2026-03-02",
        "2026-03-03",
        "2026-03-04",
        "2026-03-05"
    ]
}
df=pd.DataFrame(data)
cat=df.groupby('category')
cat 
for i,v in cat:
    print(i)
    print(v)
```

```python
#total sales of diff categories
cat=df.groupby('category')['sales'].sum()
cat
```

```python
#total for for store
store=df.groupby('store')['sales'].sum()
store
```

```python
# group by multiple column
cat=df.groupby(['category','store'])['sales'].sum()
cat
```
