## Group BY


```python
data = {
    "category": ["Electronics", "Clothing", "Groceries", "Electronics", "Electronics"],
    "store": ["Store A", "Store B", "Store C", "Store A", "Store B"],
    "sales": [15000, 8000, 3000, 22000, 12000],
    "quantity": [5, 20, 50, 8, 25],
    "date": [
        "2026-03-01",
        "2026-03-02",
        "2026-03-03",
        "2026-03-04",
        "2026-03-05"
    ]
}
df=pd.DataFrame(data)
cat=df.groupby('category')
cat 
for i,v in cat:
    print(i)
    print(v)
```

```python
#total sales of diff categories
cat=df.groupby('category')['sales'].sum()
cat
```

```python
#total for for store
store=df.groupby('store')['sales'].sum()
store
```

```python
# group by multiple column
cat=df.groupby(['category','store'])['sales'].sum()
cat
```

## Pivot table 

```python

data = {
    "Date": [
        "2025-01-01", "2025-01-02", "2025-01-03",
        "2025-01-04", "2025-01-05", "2025-01-06",
        "2025-01-07", "2025-01-08"
    ],
    "Product": [
        "Laptop", "Laptop", "Tablet", "Tablet",
        "Phone", "Phone", "Laptop", "Tablet"
    ],
    "Region": [
        "North", "South", "East", "West",
        "North", "South", "East", "West"
    ],
    "Sales": [
        1200, 1500, 800, 900,
        600, 700, 1300, 850
    ],
    "Units": [
        3, 4, 5, 6,
        10, 12, 2, 4
    ],
    "Rep": [
        "Alice", "Bob", "Charlie", "David",
        "Alice", "Bob", "Charlie", "David"
    ]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df['month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
print(df)
```

## Cross table

```python
pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='median')
```

```python
pivot2=pd.pivot_table(df,values=['Sales','Units'],index='Region',columns='Product')
```
