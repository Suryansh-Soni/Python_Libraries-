```python
import pandas as pd

# Employee DataFrame
employee = pd.DataFrame({
    "emp_id": [1, 2, 3, 4],
    "name": ["Rahul", "Priya", "Amit", "Sneha"],
    "department": ["IT", "HR", "Finance", "Marketing"]
})

# Salary DataFrame
salary = pd.DataFrame({
    "emp_id": [1, 2, 3, 4,5],
    "salary": [50000, 45000, 60000, 55000,40000],
    "bonus": [5000, 4000, 6000, 5500,4000]
})
```

```python
#merge : default inner -> based on common ids 

pd.merge(employee,salary,on='emp_id',how='inner')
```

```python
# outer->based on total number of rows adding both
pd.merge(employee,salary,on='emp_id',how='outer') 
```

```python
# left -> based on the first table ids , final no. of row == row in first dtaframe
pd.merge(employee,salary,on='emp_id',how='left')
```

```python
# right -> based on the second table ids , final no. of row == row in 2nd dtaframe
pd.merge(employee,salary,on='emp_id',how='right')
```

## concatination

```python

# First DataFrame
df1 = pd.DataFrame({
    "Name": ["Amit", "Rahul"],
    "Age": [22, 24]
})

# Second DataFrame
df2 = pd.DataFrame({
    "Name": ["Priya", "Sneha"],
    "Age": [23, 25]
})
```

```python
# default just add based on col 
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

## Joining 2 dataframe

```python
import pandas as pd

# First DataFrame
df1 = pd.DataFrame(
    {
        "names": ["suryansh",'sumit','tejas','tiwari']
    },
    index=[1, 2, 3, 4]
)

# Second DataFrame
df2 = pd.DataFrame(
    {
        "score2": [80, 89, 90, 60]
    },
    index=[2, 3, 4, 5]
)

print(df1)
print(df2)
```

```python
df1.join(df2)
```
