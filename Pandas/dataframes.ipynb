# DataFrames

## 1)creating dataframe


```python
import pandas as pd
import numpy as np
```

```python
# using dictionary
data={
    "name":['suryansh','arnab','sambhav'],
    'age':[21,20,20],
    'city':['ngp','ngp','npg']
}
df1=pd.DataFrame(data)
```

```python
# using list
data_list=[['suryansh',21,'ngp'],
           ['arnab',20,'ngp'],
           ['sambhav',20,'ngp']]
# pd.DataFrame(data_list)
columns=['name','age','city']
df2=pd.DataFrame(data_list,columns=columns)
```

## Finding Missing Data

```python
data=[['suryansh',21,np.nan],
           ['arnab',np.nan,'ngp'],
           ['sambhav',20,'ngp']]
columns=['name','age','city']
df=pd.DataFrame(data,columns=columns)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
#to check if any col do not have null value
df.isna().any()
```

```python
# row wise
df.dropna()
```

```python
# thresh : it set thresold for non null values so the row is not droped .
df.dropna(thresh=2)
```

```python
# column wise
df.fillna(0)
```

```python
values={'age':20,'city':'ngp'}
df.fillna(value=values)
```
