# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y)
plt.title("hello")
plt.xlabel("x -axis")
plt.ylabel("y-label")
```

### Subplot

```python
# subplot : divide total area in 4 area for different graph
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.plot(x,y)
```

### Object Oreanted Way 


```python
# fig.add_axes([shift from left ,from bottom to up 
# ,width size,height size])

# figsize=(weigth,height)

#dpi =dot per inch 
fig=plt.figure(figsize=(10,6),dpi=100)
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.3,0.3,0.3])
axis2.plot(x,y)

```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create parallel line data
x = np.linspace(0, 10, 10)
df = pd.DataFrame({"x": x})

# Generate parallel lines (offsets)
for i in range(1, 17):
    df[f"y{i}"] = x + i  # parallel lines

fig, ax = plt.subplots(figsize=(12, 6))

# linewidth examples
ax.plot(df["x"], df["y1"], color="red", linewidth=0.25)
ax.plot(df["x"], df["y2"], color="red", linewidth=0.50)
ax.plot(df["x"], df["y3"], color="red", linewidth=1.00)
ax.plot(df["x"], df["y4"], color="red", linewidth=2.00)

# linestyle examples
ax.plot(df["x"], df["y5"], color="green", lw=3, linestyle='-')
ax.plot(df["x"], df["y6"], color="green", lw=3, linestyle='--')
ax.plot(df["x"], df["y7"], color="green", lw=3, linestyle='-.')
ax.plot(df["x"], df["y8"], color="green", lw=3, linestyle=':')

# custom dash
line, = ax.plot(df["x"], df["y9"], color="black", lw=1.5)
line.set_dashes([5, 5, 15, 5])

# marker examples
ax.plot(df["x"], df["y10"], color="blue", lw=2, linestyle='-', marker='+')
ax.plot(df["x"], df["y11"], color="blue", lw=2, linestyle='-', marker='o')
ax.plot(df["x"], df["y12"], color="blue", lw=2, linestyle='-', marker='s')

# marker size + color
ax.plot(df["x"], df["y13"], color="purple", lw=1, marker='o', markersize=4)
ax.plot(df["x"], df["y14"], color="purple", lw=1, marker='o', markersize=8)
ax.plot(df["x"], df["y15"], color="purple", lw=1, marker='s', markersize=8, markerfacecolor="red")

# marker edge
ax.plot(df["x"], df["y16"], color="orange", lw=1, marker='o',
        markersize=8, markeredgewidth=2, markeredgecolor="green")

ax.set_title("Parallel Lines with Different Styles")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```
