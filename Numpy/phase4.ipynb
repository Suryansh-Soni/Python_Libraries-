```python
import numpy as np
import matplotlib.pyplot as plt

try:
    logo = np.load('numpylogoicon.npy', allow_pickle=True)

    plt.figure(figsize=(10,6))

    # Original image
    plt.subplot(121)
    plt.imshow(logo)
    plt.axis('off')
    plt.title("NumPy Logo")

    # Inverted image
    dark_logo = 1 - logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.axis('off')
    plt.title("Dark Logo")

    plt.show()

except FileNotFoundError:
    print("No file found.")
```
