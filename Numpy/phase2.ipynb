## Numpy Array Operations


```python
import numpy as np
```

## indexing 


```python
arr=np.arange(0,10)
arr[2:7]
print("With steps: ",arr[0:10:3])
print("neg index: ",arr[10::-1])
```

```python
arr2=np.array([[1,2,3],[4,5,6]])

arr2[1,2]
print(arr2[1]) #rows
print(arr2[:,1]) #column
```

## Sorting 

```python
arr=np.array([1,4,3,2,6,7])
print("Sroted Array:",np.sort(arr))
arr2=np.array([[1,4,3],[5,2,5]])
print("Sorted 2d:",np.sort(arr2,axis=1)) #0 for column and 1 for row
```

## Array Compatibility

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
c=np.array([5,6,7])
print("compatibality :",arr1.shape==arr2.shape)
```

```python
original=np.array([[1,2],[3,4]])
new_row=np.array([[5,6]])
#new roww
with_new_row=np.vstack((original,new_row))
print(with_new_row)
# new column
print("-----------------")
new_col=np.array([[5],[6]])
with_new_col=np.hstack((original,new_col))
print(with_new_col)
```
