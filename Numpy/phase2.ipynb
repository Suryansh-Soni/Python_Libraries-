## Numpy Array Operations


```python
import numpy as np
```

## indexing 


```python
arr=np.arange(0,10)
arr[2:7]
print("With steps: ",arr[0:10:3])
print("neg index: ",arr[10::-1])
```

```python
arr2=np.array([[1,2,3],[4,5,6]])

arr2[1,2]
print(arr2[1]) #rows
print(arr2[:,1]) #column
```

## Sorting 

```python
arr=np.array([1,4,3,2,6,7])
print("Sroted Array:",np.sort(arr))
arr2=np.array([[1,4,3],[5,7,5]])
print("Sorted 2d:",np.sort(arr2,axis=1)) #0 for column and 1 for axis
```
