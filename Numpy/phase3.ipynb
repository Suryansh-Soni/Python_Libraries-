```python
import numpy as np
import matplotlib.pyplot as plt

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=== zomato sale analysis ===")
print("\n Sales data shape",sales_data.shape)
print("\nSample data for first 3 resturents:\n",sales_data[:3])
```

```python
# Total sales per year
print(np.sum(sales_data,axis=0))
yearly_total=np.sum(sales_data[:,1:],axis=0)
yearly_total
```

```python
# min sales per restaurent
print(np.min(sales_data[:,1:],axis=0))
max_sale=np.max(sales_data[:,1:],axis=0)
print(max_sale)
avg_sale=np.mean(sales_data[:,1:],axis=1)
print(avg_sale)
```

```python
cum_sum=np.cumsum(sales_data[:,1:],axis=1)
print(cum_sum)
plt.figure(figsize=(10,6))
plt.plot(np.mean(cum_sum,axis=0))
plt.title("Avg cum_sum throughout all restaurents")
```

```python
import numpy as np

vector1 = np.array([1,2,3,4])
vector2 = np.array([6,7,8,9])

print("\nAddition vector:", vector1 + vector2)
print("\nMultiplication vector:", vector1 * vector2)
print("\nDot Product:", np.dot(vector1, vector2))

angle = (np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\nCosine of Angle:", angle)
```
