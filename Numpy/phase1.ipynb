## Array Reshaping

```python
arr=np.arange(12)
reshape=arr.reshape(3,4)
reshape
```

```python
# flating the rehaped matrix
flattened=reshape.flatten()
flattened
```

```python
# raveled:return view, insted of copy i.e return original array
raveled=reshape.ravel()
raveled
```

```python
Transpose=reshape.T
Transpose
```
